# What is our dataSet?

<h3>Description:</h3>
This dataset provides a comprehensive view of movies and shows available on Netflix. It captures a range of information, offering insights into the type of content, its popularity, and other significant aspects. This dataset is ideal for exploring trends in entertainment, understanding preferences among different age groups, and conducting comparative analysis of genres and production locations.

<h3>Features: </h3>
title: The name of the movie or show, an essential identifier.
type: Distinguishes between a movie and a show, indicating the format of the content.
release_year: The year the content was released, providing a temporal context.
age_certification: Indicates the age suitability rating, important for audience targeting.
runtime: Duration of the movie or show in minutes, reflecting content length.
genres: Categories of the content, such as comedy, drama, or horror, crucial for understanding its theme.
production_countries: The countries where the content was produced, highlighting geographical influences.
seasons: Number of seasons (applicable for shows), indicating the extent of the content.
imdb_id: A unique identifier for each title on IMDb, useful for cross-referencing.
imdb_score: The IMDb rating of the content, reflecting its popularity and quality.
imdb_votes: The number of votes the content received on IMDb, indicative of viewer engagement and popularity.
Use Case:
This dataset is invaluable for tasks like trend analysis in entertainment, predictive modeling for viewer preferences, and comparative studies of genres and release years. It allows for an in-depth understanding of the factors that contribute to the success and popularity of movies and shows on Netflix.

<h3>Note:</h3>

The dataset is a rich resource for researchers and analysts in the field of entertainment and media studies.
It can be used to correlate various features like genre, release year, and IMDb scores to understand audience preferences and trends over time.

## Data Preparation & Cleaning

While the data includes a wealth of information, we'll limit our analysis to the following areas:

- Movies released from 2000 - 2021
- imdb_id is irrelevant


Let's select a subset of columns with the relevant data for our analysis.

````python
movies_df = raw_movies_df.copy()
movies_df.drop('imdb_id', axis=1, inplace=True)
movies_df.drop('seasons',axis=1,inplace=True)
````

````python
movies_df # We  still have type == Show :( 
````

````python
movies_df = movies_df[movies_df.type=='MOVIE'] # Now we only have movies
````

````python
movies_df = movies_df[movies_df.release_year >= 2000] # Now we only have movies after and including the year 2000
````

````python
movies_df
````

````python
movies_df.shape # Our now reduced dataframe
````

````python
movies_df.info() # Column info
````

````python
# Let's handle column entries that have NaN
import numpy as np
````

````python
# Assuming 'column_name' is the name of your column
movies_df.age_certification = movies_df.age_certification.fillna("")
````

````python
movies_df # We have replace age_certification == NaN With a blank string
````

````python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
````

### Let us plot the number of movies of each year between 2000-2021 available on Netflix

````python
movies_df
````

````python
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)

# Generate a color palette with enough colors for the number of bars
num_bars = len(movies_df.release_year.value_counts().index)
color_palette = sns.color_palette("Set1", num_bars)

plt.title("Release Year vs Num of Movies Available")
plt.xlabel("Release Year")
plt.ylabel("Num of Movies Available")

sns.barplot(x=movies_df.release_year.value_counts().index,y=movies_df.release_year.value_counts(),
            palette=color_palette,
            hue=movies_df.release_year.value_counts().index,
            legend=False
           );
````

It appears that a disproportionately high number of movies on netflix are from the years 2017-2021, there can be many reasons for this, I will offer a few: 

1) Age of Viewer on Netflix
2) Movie Rights
3) Number of Movies released per Year, we would expect earlier years to have a lower number of movies released per year

We can repeat our barplot using genres top 20

````python
twentyMostCommonGenre = movies_df.genres.value_counts().head(20)
````

````python
plt.figure(figsize=(8,4))
plt.xticks(rotation=90)

# Generate a color palette with enough colors for the number of bars
num_bars = len(twentyMostCommonGenre.index)
color_palette = sns.color_palette("Set3", num_bars)

plt.title("Top 20 Movie Genre")
plt.xlabel("Genre")
plt.ylabel("Genre Count")


sns.barplot(x=twentyMostCommonGenre.index,y=twentyMostCommonGenre,
            palette=color_palette,
            hue=twentyMostCommonGenre.index,
            legend=False,
            fill=False
           );

````

WE notice our top 3 genres appear as members in almost all of the top 20
We also notice an empty Genre that appears as the top 18, this could be due to data entry error or does purposefully for movies that could not be assigned a Genre.

We now need to determine if the top 20 Genre have  an associated high IMDB_SCORE

````python
# Let us grab teh top 20 movies with the highest imdb_score and cross check their genre with the genre of the top 20
````

````python
twentyHighestScoredMovies = movies_df[movies_df.imdb_score > 0].sort_values(
    ['imdb_score', 'imdb_votes'], ascending=[False, False]
).head(20)

# we grab data using:            imdb_score imdb_votes 
````

````python
twentyHighestScoredMovies
````

````python
highestRatedGenres = twentyHighestScoredMovies.genres.value_counts()
````

````python
plt.figure(figsize=(12,6))
plt.xticks(rotation=85)

# Generate a color palette with enough colors for the number of bars
num_bars = len(highestRatedGenres.index)
color_palette = sns.color_palette("Set3", num_bars)

plt.title("Highest Rated Movies Genres ")
plt.xlabel("Genre")
plt.ylabel("Genre Count")


sns.barplot(x=highestRatedGenres.index,y=highestRatedGenres,
            palette=color_palette,
            hue=highestRatedGenres.index,
            legend=False,
            fill=True
           );
````

From this we can infer that the top 3 most common Genres "Comedy","Drama","Documentation" Do make their presence felt
in the top 20 scored movies, this trend may continue if we increase our sample space
