People often make judgments before analyzing data due to a variety of cognitive biases and heuristics that simplify decision-making processes. One of the main reasons is the **availability heuristic**, which is a mental shortcut that relies on immediate examples that come to a person's mind when evaluating a specific topic, concept, method, or decision¹. This can lead to quick judgments based on what is readily available in memory rather than on thorough data analysis.

Another reason is the **representativeness heuristic**, where people judge the probability or frequency of an event by the extent to which it resembles the typical case¹. This can result in overlooking relevant data in favor of what seems representative or familiar.

The phenomenon of making judgments without data analysis is also related to the concept of **bounded rationality**, which suggests that when individuals make decisions, their rationality is limited by the information they have, the cognitive limitations of their minds, and the finite amount of time they have to make a decision¹. As a result, people may resort to simpler and more readily accessible information rather than engaging in extensive data analysis.

These cognitive shortcuts are generally useful for navigating daily life efficiently, but they can lead to errors or biases in judgment, especially in complex situations where data analysis is required for informed decision-making¹². The study of these phenomena falls under the field of **behavioral economics** and **psychology**, particularly in the work of researchers like Daniel Kahneman and Amos Tversky, who have extensively studied how people make decisions and the biases that affect them⁴.

````python
# Visulisation Code Libraries we will be using 

import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
````

````python
plt.figure(figsize=(12,6))

# Assuming 'raw_df' is your DataFrame and 'citizenship' is the column of interest.
# First, get the counts of each unique value in the 'citizenship' column.
citizenship_counts = raw_df['citizenship'].value_counts()

# Convert the Series to a DataFrame, which is required for Seaborn.
citizenship_counts_df = citizenship_counts.reset_index()
citizenship_counts_df.columns = ['citizenship', 'counts']  # Rename columns for clarity.

# Now, create the scatter plot.
sns.scatterplot(x='citizenship', y='counts', data=citizenship_counts_df, s=300,hue=raw_df.citizenship.unique())

plt.title("Fatality by Citizenship")
# Show the plot.
plt.show()
````

````python

# Assuming 'raw_df' is your DataFrame and it's already grouped appropriately
grouped_df = raw_df.groupby(['type_of_injury', 'took_part_in_the_hostilities'])['age'].mean().unstack()


# raw_df.groupby(['type_of_injury', 'took_part_in_the_hostilities']): This part groups the DataFrame raw_df by two columns: 'type_of_injury' and 'took_part_in_the_hostilities'. It organizes the data so that rows with the same values in these columns are grouped together.

# ['age'].mean(): After grouping, this calculates the mean (average) age for each group.

# .unstack(): This method reshapes the resulting grouped data, pivoting the innermost level of the index labels to the columns. It turns the multi-level index created by groupby into a more readable format, where each unique value from the 'took_part_in_the_hostilities' column becomes a separate column in the resulting DataFrame.

# The purpose of this code is to analyze the average age of individuals based on the type of injury they sustained and their participation in hostilities, presenting the data in an organized and accessible way. If you need further analysis or visualization of this data, feel free to ask!


# Plotting the stacked bar chart
grouped_df.plot(kind='bar', stacked=True, figsize=(14,7))
plt.title("Average Age by Type of Injury and Participation in Hostilities")
plt.xlabel("Type Of Injury")
plt.ylabel("Average Age")
plt.legend(title='Took Part in the Hostilities')
plt.show()
````
