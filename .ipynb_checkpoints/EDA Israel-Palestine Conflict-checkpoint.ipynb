## Step 3: Perform exploratory analysis and Visualization

```python
# Visulization code libraries we will be using

import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style("darkgrid")
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.figsize'] = (9,5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
# I want a plot where the x-axis is the citizenship and the y-axis is the number of fatalaties
raw_df.citizenship.unique()
```

```python
plt.figure(figsize=(12,6))

citizenship_counts = raw_df['citizenship'].value_counts()

citizenship_counts_df = citizenship_counts.reset_index()
citizenship_counts_df.columns = ['citizenship','counts']

sns.scatterplot(x='citizenship',y='counts',data=citizenship_counts_df,s=300,hue=raw_df.citizenship.unique())

plt.title("Fatality by Citizenship")

plt.show();
```

<p>If we analyse the scatterplot, we confirm that Palestine suffered far more Fatalities as compared to their Israeli Counterparts.</p>

<p> <bold> This is quite interesting, western media sources tend to portray Palestine and especially the deemed "Terrorist" group Hamas
as a Brutal aggressors, and yet the data does not substantiate these accusations.</bold> </p> 

```python
plt.title("Fatality vs Citizenship")
plt.xlabel("Citizenship")
plt.ylabel("Num. of Fatalities")

plt.hist(raw_df.citizenship,color='r');
```

```python
raw_df_copy = raw_df.copy()

raw_df_copy['year_of_event'] = pd.to_datetime(raw_df_copy['date_of_event']).dt.year
raw_df_copy['month_of_event'] = pd.to_datetime(raw_df_copy['date_of_event']).dt.month
```

```python
raw_df_copy
```

```python
# I want to view the fatalities per month over the years


heatmap_data = raw_df_copy.pivot_table(index="month_of_event",columns="year_of_event",values="name",aggfunc="count")

plt.figure(figsize=(12,8))
sns.heatmap(heatmap_data,cmap="Reds",annot=True,fmt='g')
plt.title("Fatalities Heatmap")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show();
```

## Darker Hues

1) Dec 2008 - Jan 2009: This increase in fatalies is attributed to the Gaza war
2) July 2014 - August 2014: Attributed to Operation Protective edge
3) The obvious conclusion more death during war times.
